```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10,100)      #np.linspace(start, stop, num)
plt.plot(x,np.sin(x),'g-')  #g- makes green solid line as g indicates green and - indicates solid line
plt.plot(x,np.cos(x),'r--')   #same r indicate red line and -- indicates dotted line
plt.show()
```

```python
#Bar Plot
fig=plt.figure() # create a figure
ax=fig.add_axes([0,0,1,1]) #add axes
langs=['C','C++','Java','Python','PHP'] #same labels as above
students=[23,17,35,29,12] #same data as above
ax.bar(langs,students) # plot a bar graph of lang and student
plt.xlabel('Language') #X-axis label
plt.ylabel('Number of people') #y-axis label
plt.show()
```

```python
# Pie Chart
import matplotlib.pyplot as plt
fig1=plt.figure()#create a new figure
ax=fig1.add_axes([0,0,1,1])#add axes to the figure
langs=['C','C++','Java','Python','PHP']
students=[23,17,35,29,12]
ax.pie(students,labels=langs,autopct='%1.2f%%') #autopct='%1.2f%%'  Displays percentage values with two decimal places.
plt.show()
```

```python
#Scatter Plot
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

```python
#3D Scatter Plot
fig3=plt.figure()
ax=plt.axes(projection='3d')#3d plot karana hai
z=20*np.random.random(100)#isma 0-1 kai bich mai random 100 number aayenga or fir voh 20 sai multiply honga toh voh 0-10 kai bich mai aa jayega
x=np.sin(z)
y=np.cos(z)
ax.scatter3D(x,y,z,c=z,cmap='Greens')
plt.show()
```

```python
#house price california
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
data=fetch_california_housing()
datas= pd.DataFrame(data.data, columns=data.feature_names)
datas['price']=data.target
print(datas.head())
sns.histplot(datas['price'],kde=True)
plt.show()
correlation=datas.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.4f',annot=True,annot_kws={'size':8},cmap='Blues')
plt.show()
```
